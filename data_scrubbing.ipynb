```python
import pandas as pd
```

```python
df = pd.read_csv('listings.csv')
```

```python
df
```

```python
df.columns
```

```python
columns = ['id', 'name', 'host_id', 'host_name', 'host_is_superhost',
                   'neighbourhood', 'price',
                   'beds', 'review_scores_rating']
```

```python
df = df[columns]
```

```python
df.dropna(inplace=True)
```

```python
len(df)
```

```python
df['neighbourhood'].unique()
```

```python
df['neighbourhood'] = df['neighbourhood'].replace(['San Francisco, California, United States', 'san francisco, California, United States'], 'San Francisco')
df['neighbourhood'] = df['neighbourhood'].replace('San Francisco, Hayes Valley, California, United States', 'Hayes Valley')
df['neighbourhood'] = df['neighbourhood'].replace('Noe Valley - San Francisco, California, United States', 'Noe Valley')
```

```python
df['neighbourhood'].unique()
```

```python
df.dtypes
```

```python
df['price'] = df['price'].str.replace('$', '')
df['price'] = df['price'].str.replace(',', '')
df['price'] = df['price'].astype(float)
```

```python
df.dtypes
```

```python
cleaned_df = df
```

```python
cleaned_df
```

```python
cleaned_df.to_csv('listings_clean.csv', index=False)
```
